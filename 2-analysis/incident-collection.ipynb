### Data Processing

#### AIAAIC Repository

```python
# Remove unnecessary columns from AIAAIC dataframe
aaic_columns = ['AIAAIC ID#', 'Headline', 'Occurred',
                'Country(ies)', 'Sector(s)', 'Deployer(s)',
                'Developer(s)', 'System name(s)', 'Technology(ies)',
                'Summary/links']

aaic_df = aaic_df[aaic_columns]

aaic_df = aaic_df.rename(columns={
    'AIAAIC ID#': 'Incident_ID',
    'Occurred': 'Date',
    'Country(ies)': 'Country',
    'Sector(s)': 'Sector', 
    'Deployer(s)': 'Deployer',
    'Developer(s)': 'Developer',
    'System name(s)': 'System',
    'Technology(ies)': 'Technology',
    'Summary/links': 'Link'
})
```

```python
# Remove rows with all NA values
aaic_df = aaic_df.dropna(how='all').reset_index(drop=True)
```

```python
# Set column types
print(aaic_df['Date'].dtype)
print(aaic_df['Country'].dtype)
print(aaic_df['Sector'].dtype)
print(aaic_df['Deployer'].dtype)
print(aaic_df['Developer'].dtype)
print(aaic_df['System'].dtype)
print(aaic_df['Technology'].dtype)
print(aaic_df['Link'].dtype)

# Convert all columns to string
aaic_df = aaic_df.astype('string')

# Extract start date from ranges
aaic_df['Date'] = aaic_df['Date'].str.extract(r'(\d{4})')[0].astype('Int64')

aaic_df.head()
```

```python
# Regex for filtering
deepfake_regex = re.compile(
    r'(deep[\s\-]?fake[s]?|voice (recreation|generation)|generative adversarial network|AI[-\s]?generated|fake (image|video|audio))',
    flags=re.IGNORECASE
)

# Filter by Headline and Technology columns by regex
aaic_deepfakes_df = aaic_df[
    aaic_df['Headline'].str.contains(deepfake_regex, na=False) |
    aaic_df['Technology'].str.contains(deepfake_regex, na=False)
]

# Count deepfake related incidents
aaic_deepfake_count = aaic_deepfakes_df.shape[0]

print(f"Deepfake related incidents: {aaic_deepfake_count}")
```

```python
# Subset data -- this will go later
df_subset = aaic_deepfake_df[['Incident_ID', 'Date', 'Sector']].copy()

# Expand sectors into dummy variables
sector_dummies = df_subset['Sector'].str.get_dummies(sep=';')

# Clean the column names (remove leading/trailing spaces)
sector_dummies.columns = sector_dummies.columns.str.strip()

# Group the cleaned column names and sum the values
sector_dummies = sector_dummies.groupby(sector_dummies.columns, axis=1).sum()

# Combine
df_sector = pd.concat([df_subset, sector_dummies], axis=1)

# Combine related columns
df_sector['Personal'] = df_sector['Personal - individual'] + df_sector['Private - individual, family']
df_sector['Defence'] = df_sector['Govt - defence'] + df_sector['Govt - military']
df_sector['Police'] = df_sector['Govt - police'] + df_sector['Govt - security']

# Drop original columns
df_sector = df_sector.drop(columns=['Sector',
                                    'Personal - individual',
                                    'Private - individual, family',
                                    'Govt - defence',
                                    'Govt - military',
                                    'Govt - police',
                                    'Govt - security'])

# Rename columns
df_sector = df_sector.rename(columns={
    'Banking/financial services': 'Financial services',
    'Business/professional services': 'Professional services',
    'Govt': 'Government',
    'Govt - finance': 'Tax',
    'Govt - foreign': 'Foreign affairs',
    'Govt - justice': 'Justice',
    'Govt - telecoms': 'Telecommunications',
    'Media/entertainment/sports/arts': 'Media & Entertainment',
    'Transport/logistics': 'Transport & Logistics',
    'Travel/hospitality': 'Travel & Hospitality'
})
```

```python
# Goal, Methods, and Failures Filtering: https://incidentdatabase.ai/taxonomies/gmf/
aiid_gmf_deepfake_keys = ['deepfake video generation',
                          'voice generation',
                          'deepfake image generation']

# Transform Known AI Goal column into lowercase for consistency, fill NA with empty string
aiid_class_df['ai_goal_lower'] = aiid_class_df['Known AI Goal'].fillna("").astype(str).str.lower()

# Split comma separated values for consistency
aiid_class_df['ai_goal_list'] = aiid_class_df['ai_goal_lower'].str.split(",").apply(
    lambda x: [goal.strip() for goal in x if goal.strip()]  # skip empty strings
)

# Filter rows with deepfake-related AI goals
aiid_class_deepfakes_df = aiid_class_df[aiid_class_df["ai_goal_list"].apply(
    lambda goals: any(goal in aiid_gmf_deepfake_keys for goal in goals)
)]

aiid_class_deepfakes_df
```

### Web Scraping

```python
# Import libraries
import time
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
from urllib.parse import urljoin
import re
from datetime import datetime
```

```python
# Regex to detect URLs in plain text
URL_PATTERN = re.compile(
    r'https?://[^\s)]+',
    re.IGNORECASE
)

# Define function to retrieve news articles
def get_article_links(soup, base_url=None):
    """
    Extracts article links or text references from relevant sections on the page.
    Sections include:
      - News / Commentary / Analysis
      - Fact check
      - Investigations / Assessments / Audits
    Deduplicates results while preserving order.
    Returns: list of (url, section_type)
    """

    section_map = {
        "news": "News / Commentary / Analysis",
        "commentary": "News / Commentary / Analysis",
        "analysis": "News / Commentary / Analysis",
        "fact check": "Fact Check",
        "investigations": "Investigations / Assessments / Audits",
        "assessments": "Investigations / Assessments / Audits",
        "audits": "Investigations / Assessments / Audits"
    }

    def header_matches(tag):
        if tag.name not in ("h3", "span"):
            return False
        text = tag.get_text(" ", strip=True).lower()
        return any(keyword in text for keyword in section_map)

    all_links = []
    seen = set()  # to track duplicates

    for header_tag in soup.find_all(header_matches):
        h3 = header_tag if header_tag.name == "h3" else header_tag.find_parent("h3")
        if not h3:
            continue

        # Identify which section this header corresponds to
        header_text = h3.get_text(" ", strip=True).lower()
        section_type = None
        for keyword, label in section_map.items():
            if keyword in header_text:
                section_type = label
                break
        if not section_type:
            section_type = "Other"

        # Find the next <ul> until another <h3>
        ul = None
        for tag in h3.next_elements:
            if tag.name == "h3":
                break
            if tag.name == "ul":
                ul = tag
                break
        if not ul:
            continue

        # Go through everything inside the <ul>
        for descendant in ul.find_all(["a", "p", "span"]):
            if descendant.name == "a" and descendant.has_attr("href"):
                # Case 1: real hyperlink
                href = descendant["href"].strip()
                if href.startswith(("#", "mailto:", "javascript:")):
                    continue
                if base_url:
                    href = urljoin(base_url, href)
                if href not in seen:
                    all_links.append((href, section_type))
                    seen.add(href)
            else:
                # Case 2: plain text URL
                text = descendant.get_text(" ", strip=True)
                matches = URL_PATTERN.findall(text)
                for match in matches:
                    if match not in seen:
                        all_links.append((match, section_type))
                        seen.add(match)

    return all_links
```

```python
# --- setup requests session ---
session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (compatible; DeepfakeScraper/1.0; +https://github.com/treinmund/deepfake-threatlandscape)"
})
```

```python
all_records = []

for _, row in tqdm(deepfake_incidents.iterrows(), total=len(deepfake_incidents)):
    incident_id = row["Incident_ID"]
    summary_url = row["Link"]

    try:
        response = requests.get(summary_url, timeout=15)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        links = get_article_links(soup, base_url=summary_url)

        for idx, (url, section_type) in enumerate(links, start=1):
            all_records.append({
                "Incident_ID": incident_id,
                "Article_Number": idx,
                "URL": url,
                "Section_Type": section_type
            })
        time.sleep(0.5)  # polite delay
    
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {summary_url}: {e}")

# Build dataframe
df_articles = pd.DataFrame(all_records)

# Get current date and time
now = datetime.now()
timestamp = now.strftime("%Y%m%d_%H%M%S")

# Create filename with timestamp
filename = f"article_links_{timestamp}.csv"

# Save to CSV
output_dir = "/Users/tylund/Library/CloudStorage/Dropbox/1. Side Projects/2025.1-Deepfake Threat Landscape/1-data/web-data/"
path_articles = os.path.join(output_dir, filename)
df_articles.to_csv(path_articles, index=False)

print(f"Scraping complete. Saved to {filename}")
```
